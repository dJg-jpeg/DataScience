```python
from tensorflow import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.optimizers import Adadelta
from keras.activations import softmax, relu, selu
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

class_names = ['T-shirt', 'Trousers', 'Pullover', 'Dress', 'Coat', 
               'Sandals', 'Shirt', 'Sneakers', 'Bag', 'Ankle boots']

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
neural_net = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(1024, activation=relu),
    Dense(512, activation=selu),
    Dense(10, activation=softmax),
])

neural_net.compile(
    optimizer=Adadelta(learning_rate=0.01),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
)


x_val = x_train[:10000]
pt_x_train = x_train[10000:]
y_val = y_train[:10000]
pt_y_train = y_train[10000:]

history = neural_net.fit(pt_x_train, pt_y_train, epochs=200, batch_size=256, validation_data=(x_val, y_val))
```

```python
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

epochs = range(1, len(history_dict['accuracy']) + 1)

plt.plot(epochs, loss_values, 'yo-', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.clf()
val_acc_values = history_dict['val_accuracy']
plt.plot(epochs, history_dict['accuracy'], 'yo-', label='Training acc')
plt.plot(epochs, history_dict['val_accuracy'], 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
```

```python
results = neural_net.evaluate(x_test, y_test)
print(results)
```
