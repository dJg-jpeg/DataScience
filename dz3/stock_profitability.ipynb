```python
companies_data = pd.read_csv("companies.csv")
companies_data
```

```python
sns.regplot(x="a", y="b", data=companies_data).set(title="Графік кореляції між компаніями А та В")
```

```python
sns.regplot(x="b", y="c", data=companies_data).set(title="Графік кореляції між компаніями B та C")
```
