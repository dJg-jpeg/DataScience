```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
juniors = pd.read_csv("https://drive.google.com/u/0/uc?id=1JMYqXipZpz9Y5-vyxvLEO2Y1sRBxqu-U&export=download")
juniors.drop(["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1, inplace=True)
juniors.dropna(inplace=True)
```

```python
python_juns = juniors[juniors["Язык.программирования"] == "Python"]
position = python_juns.groupby("Должность")
position = position["Зарплата.в.месяц"].agg("mean")
plt.xlabel("Посада")
plt.ylabel("Зарплата")
plt.title("Порівняння середньої з/п Python розробників по посадах")
plt.xticks(rotation=15)
plt.bar(position.keys(), position.values)
plt.show()
```

```python
by_lang = juniors.groupby("Язык.программирования")
by_lang = by_lang["Зарплата.в.месяц"].agg(["min", "max", "mean"])
plt.xlabel("Мова програмування")
plt.ylabel("Зарплата")
plt.grid()
plt.title("Мінімальна , максимальна і середня з/п розробників по мовам програмування")
plt.plot(by_lang, label=["min", "max", "mean"])
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
