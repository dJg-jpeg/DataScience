```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw_triangle(apexes):
    x, y = sorted([random.random(), random.random()])
    s = x
    t = y - x
    u = 1 - y
    needed_dot = (
        s * apexes[0][0] + t * apexes[1][0] + u * apexes[2][0], 
        s * apexes[0][1] + t * apexes[1][1] + u * apexes[2][1]
    )
    dot_to_display = []
    for _ in range(50000):
        random_apex = random.choice(apexes)
        apex_x = random_apex[0]
        apex_y = random_apex[1]
        middle = ((needed_dot[0] + apex_x) / 2, (needed_dot[1] + apex_y) / 2)
        needed_dot = middle
        dot_to_display.append(middle)

    x = [x for x, y in apexes] + [x[0] for x in dot_to_display]
    y = [y for x, y in apexes] + [y[1] for y in dot_to_display]
    plt.scatter(x, y)


draw_triangle([(0.0, 0.0), (2.0, 4.0), (4.0, 0.0)])
```
